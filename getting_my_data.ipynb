```python
import cv2 as cv
import numpy as np
import os
```

```python
data_path = "data/my_data_in_progress/"
i = 0
```

```python
capture = cv.VideoCapture(0)
```

```python
while True:
    isTrue, frame = capture.read()
    
    frame = cv.resize(frame, (1920, 1080))
    
    cv.imwrite(os.path.join(data_path, str(i) + ".jpg"), frame)
    i += 1
    
    cv.imshow("VIDEO", frame)
    
    if cv.waitKey(500) & 0xFF == ord("q"):
        break

capture.release()
cv.destroyAllWindows()
```
